```
#Importing packages
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import warnings

warnings.filterwarnings("ignore")
```

```
#Importing the data
data = pd.read_csv("train.csv")
data.head(10)
```

```
#checking the shape of the data
data.shape
```

```
#splitting independent and dependent variables
X = data.drop('loan_status',1)
X.head(5)
```

```
y = data[["loan_status"]].copy()
y.head(5)
```

```
#standardizing the data
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```
# split the dataset
X_train , X_val, y_train , y_val = train_test_split(X,y, test_size=0.2, random_state =7)

y_train = pd.get_dummies(data = y_train, columns = y_train.columns)
y_val = pd.get_dummies(data = y_val, columns = y_val.columns)
```

```
# import libraries
from keras import models
from keras import layers 

# sequential model
model = models.Sequential()

# input layer
model.add(layers.Dense(128,input_dim = X_train.shape[1], activation = 'relu'))

# hidden layer
model.add(layers.Dense(128, activation = 'relu'))

# hidden layer
model.add(layers.Dense(64, activation = 'relu'))

# hidden layer
model.add(layers.Dense(10, activation = 'softmax'))

# compile the model
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

history = model.fit(X_train, y_train, batch_size = 64, epochs = 5)

#calculate the accuracy
accuracy = model.evaluate(X_val,y_val)
print(accuracy[1])
```

```
#predictions
# load the test data
test_data = pd.read_csv('test.csv')
test_data.head(10)
```

```
#standardize the data
test_data = scaler.transform(test_data)

# predictions
ss = pd.DataFrame(model.predict(test_data))

# create the submission file
submission=ss.idxmax(1)
submission.to_csv('Submission.csv',index=False,header=['loan_status'])
```
